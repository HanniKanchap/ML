```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
```

```python
## Create dataset
X,y = make_classification(n_samples=1000,n_features=10,n_classes=2,random_state=10)
```

```python
X = pd.DataFrame(X)
```

```python
X.head(4)
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test ,y_train , y_test = train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression()
log_reg.fit(X_train,y_train)
```

```python
y_pred = log_reg.predict(X_test)
```

```python
y_pred
```

```python
## Probability that a test data belongs to cat1 , cat2
log_reg.predict_proba(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

### Hyperparameter tuning and cross validation

```python
model = LogisticRegression()
penalty  = ['l1','l2','elasticnet']
c_values = [100,10,1.0,0.1,0.01]
solvers = ['newton-cg','lbfgs','sag','saga','liblinear']
```

```python
params = dict(penalty = penalty,C = c_values,solver = solvers)
```

```python
from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold(5)
```

```python
## GridSearchCV
from sklearn.model_selection import GridSearchCV

grid = GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=cv,n_jobs=-1)
```

```python
print(grid)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_p = grid.predict(X_test)
```

## Randomised Search CV

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
randomcv = RandomizedSearchCV(estimator=model,param_distributions=params,cv = 5,scoring='accuracy')
```

```python
randomcv.fit(X_train,y_train)
```
