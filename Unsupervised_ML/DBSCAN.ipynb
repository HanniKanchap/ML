```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X,y = make_moons(n_samples=250,noise=0.05,random_state=42)
```

```python
plt.scatter(X[:,0],X[:,1],c= y)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
model = DBSCAN(eps = 0.3)
```

```python
model.fit(X_scaled)
```

```python
model.metric
```

```python
plt.scatter(X_scaled[:,0],X_scaled[:,1],c = model.labels_)
```

##### DBSCAN Anomaly detection

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X,y = make_circles(n_samples=750,noise=0.1,factor = 0.3,random_state=42)
```

```python
X
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
from sklearn.cluster import DBSCAN
```

```python
model = DBSCAN(eps=0.10)
```

```python
model.fit_predict(X)
```

```python
import numpy as np
index = np.where(model.labels_ < 0)
print(index)
```

```python
plt.scatter(X[:,0],X[:,1],c=model.labels_)
plt.scatter(X[index,0],X[index,1],edgecolors='r',c='m')
```

```python
## Outlier data points
import pandas as pd
df = pd.DataFrame(X)
```
