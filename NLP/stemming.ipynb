```python
words = ['eaten','eat','program','programming','writing','writes','python','pythonic','finally','final','finalize','ate','history','historical','happily','happiness','happier','happiest']
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
for w in words:
    print(w,'--->',ps.stem(w))
```

```python
from nltk.stem import RegexpStemmer
```

```python
reg_stemmer = RegexpStemmer('ing$|s$|e$|able$',min = 4)
```

```python
reg_stemmer.stem('eats')
```

```python
reg_stemmer.stem('eaten')
```

```python
from nltk.stem import SnowballStemmer
```

```python
stemmer = SnowballStemmer('english')
```

```python
for w in words:
    print(w,'--->',stemmer.stem(w))
```

```python
stemmer.stem('fairly'),ps.stem('fairly')
```

```python
stemmer.stem('suprisingly'),ps.stem('suprisingly')
```
