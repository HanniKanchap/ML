```python
words = ['eaten','eat','program','programming','writing','writes','python','pythonic','finally','final','finalize','ate','history','historical','happily','happiness','happier','happiest']
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
'''POS : Noun - n , verb - v,adjective - a ,adverb - r'''
lemmatizer.lemmatize('goes')
```

```python
lemmatizer.lemmatize('going')
```

```python
lemmatizer.lemmatize('going',pos='v')
```

```python
for w in words:
    print(w,'-->',lemmatizer.lemmatize(w,pos = 'v'))
```

```python
lemmatizer.lemmatize('fairly',pos ='v')
```
